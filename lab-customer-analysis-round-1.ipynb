```python
#Standarizing the names of the columns, also changing the name of one of the colummns in File 3

f1.columns = f1.columns.str.lower().str.replace(" ", "_")
f2.columns = f2.columns.str.lower().str.replace(" ", "_")
f3.columns = f3.columns.str.lower().str.replace(" ", "_")

f3 = f3.rename(columns={'state':'st'})
```

```python
f1_order = ['customer', 'gender', 'education','vehicle_class', 'st', 'income','policy_type', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']
ord_f1 = f1[f1_order]
f2_order = ['customer', 'gender', 'education','vehicle_class', 'st', 'income','policy_type', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']
ord_f2 = f2[f2_order]
f3_order = ['customer', 'gender', 'education','vehicle_class', 'st', 'income','policy_type', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']
ord_f3 = f3[f3_order]
```

```python
#Concatenating the three files.
data = pd.concat([ord_f1, ord_f2, ord_f3]).reset_index(drop=True)
data
```

```python
###Column Types

###For the columns we have on the three files, the types are as folow:
    
    #-customer                      Categorical -- This column gives us the ID of every customer.
    #-st                            Categorical -- In this column we have the state of each customer
    #-gender                        Categorical --This column represents the gender of each customer
    #-education                     Categorical --Here we have the education level of every customer
    #-customer_lifetime_value       Numerical  -- Over here we have the value of the customer over the lifetime
    #-income                        Numerical -- Here we have the income of every customer
    #-monthly_premium_auto          Numerical -- 
    #-number_of_open_complaints     Numerical -- Here is the number of complaints that each customer has open
    #-policy_type                   Categorical -- Here is the type of policy thaht customer has contracted.
    #-vehicle_class                 Categorical -- In this column we have which type of vehicle the customer has
    #-total_claim_amount            Numerical -- Here is the total amount of money for all the complaints open for each customer
```

```python
#Deleting the columns education and number of open complaints
data = data.drop('education', axis=1)
data = data.drop('number_of_open_complaints', axis=1)
```

```python
#Correcting the values of the column customer_lifetime_value

data['customer_lifetime_value'] = data['customer_lifetime_value'].str.rstrip('%').astype(float) * 100
data
```

```python
#Checking for duplicates and deleting them if there are any
data.drop_duplicates()
```

```python
#Filtering and printing the customers that have a value of 0 or less in income.

income_filter = data[(data['income'] == 0)]
print(income_filter)
```
